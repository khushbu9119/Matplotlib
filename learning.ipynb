UNIVARTIATE - NUMERICAL

```python
Data={
 'salary':[24000,	25000,27000,30000,35000,77000,80000,90000,100000,120000,45000,50000,61000,78000,80000,90700,10000,12000,15000,20000]
}
```

```python
df=pd.DataFrame(Data)
df
```

```python
df.shape
```

```python
plt.plot(df['salary'],color='pink',marker='*',linestyle='dashed',linewidth=2,markersize=10)

plt.show()
```

```python
# histogram chart
plt.hist(df['salary'],bins=5,color='pink' )
plt.show()
```

```python
#boxplot chart
plt.boxplot(df['salary'])
plt.show()
```

```python
df
df['dept']=['HR','IT','IT','HR','IT','Finance','HR','IT','Finance','HR','IT','Finance','IT','IT','Finance','HR','IT','Finance','HR','IT']
```

```python

counts=df['dept'].value_counts()
print(counts)
```

```python
plt.pie(counts,labels=counts.index,colors=['pink','lightblue','lightgreen'],autopct='%1.1f%%',explode=[0.1,0.1,0.1],shadow=True,startangle=90)
plt.show()
```

BIVARIATE:- NUMERICAL- CATEGORICAL

```python
hr_sal=df[df['dept']=='HR']['salary']
print(hr_sal)
```

```python
it_sal=df[df['dept']=='IT']['salary']
print(it_sal)
```

```python
fin_sal=df[df['dept']=='Finance']['salary']
print(fin_sal)
```

```python
#boxplot
plt.boxplot([it_sal,hr_sal,fin_sal],tick_labels=['IT','HR','Finance'])
plt.show()
```

```python
#pie chart
salary_by_dept=df.groupby('dept')['salary'].sum()
salary_by_dept
```

```python
plt.pie(salary_by_dept,colors=['lightgreen','pink','lightblue'],labels=salary_by_dept.index,autopct='%1.1f%%',shadow=10,startangle=90,explode=[0.01,0.01,0.01])
plt.show()
```

```python
#bar plot
plt.bar(salary_by_dept.index,salary_by_dept,color=['brown','gray','black'] )
plt.show()
```

MULTIVRIATE ANALYSIS: 3 NUMERICAL COLUMNS

```python
df['Experince']=[2,3,4,5,6]*4
df
```

```python
#bubble plot
plt.scatter(df['Age'],df['salary'],s=df['Experince']*20,color='lightblue',edgecolors='black' )
plt.title('Age VS Salary VS Experince')
plt.xlabel('Age')
plt.ylabel('salary')
plt.show()
```

MULTIVRIATE ANALYSIS: 2 NUMERICAL COLUMNS 1 CATEGORICAL COLUMN

```python
plt.scatter(df['Age'],df['salary'],c=df['dept'].map({'HR':'yellow','IT':'green','Finance':'red'}))
plt.xlabel('Age')
plt.ylabel('salary')
plt.show()
```
